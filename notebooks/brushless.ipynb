```python
import math
```

```python
shaft_diameter=6
shaft_l=20
shaft_base_d=20
shaft_base_l=8
engine_l=47.2
engine_d=35
rear_shaft_l=17.3
rear_shaft_d=5.0
#total_length=
```

```python
engine_body = cq.Workplane("YZ").circle(shaft_diameter/2).extrude(-shaft_l)\
    .translate((-shaft_base_l,0,0))\
    .faces(">X").workplane().circle(shaft_base_d/2).extrude(shaft_base_l).tag("nose")\
    .edges(">X").chamfer(shaft_base_l*0.99,shaft_base_l*0.25)\
    .faces(">X").circle(engine_d/2).extrude(engine_l)\
    .edges(">X").chamfer(engine_l*0.1)\
    .faces(">X[-2]").edges().first().chamfer(engine_l*0.1).tag("tail")\
    .faces(">X").circle(rear_shaft_d/2).extrude(rear_shaft_l)
engine_body
```

```python
inner_screw_hole_cirle_A_d = 25
inner_screw_hole_circle_B_d = 19
outer_screw_hole_circle_d = 34
screw_hole_d = 3.2
thickness_l = 4
```

```python
mounting_plate = cq.Workplane("YZ")\
.rect(outer_screw_hole_circle_d+2*screw_hole_d,2*screw_hole_d).extrude(thickness_l)\
.edges("|X").fillet(screw_hole_d*.999)\

mounting_plate = mounting_plate.rotate(cq.Vector(0, 0, 0),cq.Vector(10, 0, 0),90)\
.union(mounting_plate).rotate(cq.Vector(0, 0, 0),cq.Vector(10, 0, 0),90)\
.rotate(cq.Vector(0, 0, 0),cq.Vector(10, 0, 0),45).faces(">X")\
.rect(math.sqrt(1/2)*outer_screw_hole_circle_d,math.sqrt(1/2)*outer_screw_hole_circle_d, forConstruction=True)\
.vertices().cskHole(screw_hole_d, screw_hole_d*1.9, 82, depth=None)\
.edges(cq.NearestToPointSelector((0,0,screw_hole_d))).fillet(screw_hole_d*.999)\
.edges(cq.NearestToPointSelector((0,0,-screw_hole_d))).fillet(screw_hole_d*.999)\
.edges(cq.NearestToPointSelector((0,screw_hole_d,0))).fillet(screw_hole_d*.999)\
.edges(cq.NearestToPointSelector((0,-screw_hole_d,0))).fillet(screw_hole_d*.999)\

mounting_plate

```

```python
engine_ass = cq.Assembly(engine_body, cq.Location(cq.Vector(-shaft_base_l,0,0)), name="root")
engine_ass.add(mounting_plate, loc=cq.Location(cq.Vector(engine_l,0,0)), name="mount")
engine_ass
```

```python
engine_ass.save('engine.step')
```
