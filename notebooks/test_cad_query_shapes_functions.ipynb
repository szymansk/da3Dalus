```python
rectanglePoints = ((-2,0),(0,2),(2,0),(0,-2),(-2,0))
starPoints = ((-2,0),(-3,3),(0,2),(3,3),(2,0),(3,-3),(0,-2),(-3,-3),(-2,0))
starPoints1 = ((-3,0),(-2,2),(0,3),(2,2),(3,0),(2,-2),(0,-3),(-2,-2),(-3,0))

rootsegment = cq.Workplane().polyline(starPoints).close()
segment1 = rootsegment.transformed(offset=[0,0,5]).polyline(starPoints1).close().loft()
# segment1.loft() equals section1,
# which describes the conntection from rootsegment to segment1
# last segment needs .loft()
segment2 = segment1.wire(">Z").transformed(offset=[0,0,10]).polyline(rectanglePoints).close()

display(segment1)
```

```python
rectanglePoints = ((-2,0),(0,2),(2,0),(0,-2),(-2,0))
starPoints = ((-2,0),(-3,3),(0,2),(3,3),(2,0),(3,-3),(0,-2),(-3,-3),(-2,0))
starPoints1 = ((-3,0),(-2,2),(0,3),(2,2),(3,0),(2,-2),(0,-3),(-2,-2),(-3,0))

rootsegment = cq.Workplane().polyline(starPoints).close()
segment1 = rootsegment.transformed(offset=[0,0,5]).polyline(starPoints1).close()
segment2 = segment1.wire(">Z").transformed(offset=[0,0,10]).polyline(rectanglePoints).close().loft()
# segment2.loft() equals section1 and section 2 due to additive behaviour,
# which describes the conntection from rootsegment to segment2
# last segment needs .loft()

display(segment2)
```

```python
# X axis line length 20.0
path = cq.Workplane("XZ").moveTo(-10, 0).lineTo(10, 0)

# Sweep a circle from diameter 2.0 to diameter 1.0 to diameter 2.0 along X axis length 10.0 + 10.0
defaultSweep = (
    cq.Workplane("YZ")
    .workplane(offset=-10.0)
    .circle(2.0)
    .workplane(offset=10.0)
    .circle(1.0)
    .workplane(offset=10.0)
    .circle(2.0)
    .sweep(path, multisection=True)
)

display(defaultSweep)
```
