```python
import math

# Konstante für die Umrechnung von N in g
N_TO_G = 0.101971621

# KV-Zahl des Motors
kv = 1500

# Anzahl der zellen
cells = 6

# Spannung des Lipo-Akkus
voltage = 3.7 * cells

# Leistung des Motors in Watt
power = 200

# Durchmesser der Propeller in Zoll
prop_diameter = 3.5

# Pitch der Propeller in Zoll
prop_pitch = 3

# Luftdichte in kg/m^3 (Standard: 1.225 kg/m^3)
rho = 1.225

# Umdrehungen pro Minute berechnen
rpm = kv * voltage

# Schub berechnen
thrust = (power / rpm) * N_TO_G

# Fläche der Propeller berechnen
prop_area = math.pi * (prop_diameter / 2) ** 2

# Schubdichte berechnen
thrust_density = thrust / prop_area

# Ausgabe
print("KV-Zahl: ", kv)
print("Spannung: ", voltage, "V")
print("Leistung: ", power, "W")
print("Propellerdurchmesser: ", prop_diameter, "Zoll")
print("Propellerpitch: ", prop_pitch, "Zoll")
print("Umdrehungen pro Minute: ", rpm)
print("Schub: ", thrust, "g")
print("Schubdichte: ", thrust_density, "g/cm^2")
```
